```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['Train_loss'], label='Train Loss', color='blue')

plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.savefig('loss_plot.png')
```
